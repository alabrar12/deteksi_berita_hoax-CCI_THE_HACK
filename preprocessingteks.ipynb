# PREPOSESING

```python
import re
```

```python
def preprose_teks(teks):
    if not isinstance(teks, str):
        teks = ""
    
    teks = teks.lower() #lowercase
    teks = re.sub(r"[^a-z0-9\s-]", " ", teks) #karakter non-alfanumerik
    teks = re.sub(r"\s+", " ", teks).strip() #hapus spasi berlebihan
    tokens = teks.split() #Tokenize
    clean_text = " ".join(tokens)
    
    return clean_text, tokens
```

```python
df_combin["clean_text"], df_combin["tokens"] = zip(*df_combin["text_new"].apply(preprose_teks))
```

```python
data = pd.DataFrame(df_combin['clean_text'])
data = data.rename(columns={'clean_text': 'teks_bersih'})
```

```python
data['token'] = df_combin["tokens"]
```
